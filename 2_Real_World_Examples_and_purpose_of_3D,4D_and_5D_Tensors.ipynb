````python
#Visualizing the above tensor using plotly google colab provided plotting lib
import plotly.express as px
import numpy as np

data = threeD_array

# Create meshgrid for 3D plot
x = np.arange(0, data.shape[0])
y = np.arange(0, data.shape[1])
x, y = np.meshgrid(x, y)
fig = px.scatter_3d(x=x.flatten(), y=y.flatten(), z=data.flatten(), opacity=0.7)
fig.update_traces(marker=dict(size=5))
fig.show()
````


**Sample Points:**

In the context of audio processing, a "sample" refers to a measurement of the amplitude of the audio signal at a specific point in time. Audio signals are continuous in nature, but for digital processing, they need to be discretized. Sampling involves taking measurements (samples) of the audio waveform at regular intervals.

The number of samples per second is known as the "sampling rate." For example, a common sampling rate is 44.1 kHz (44,100 samples per second). If you have a 10-second audio clip recorded at a 44.1 kHz sampling rate, you would have 441,000 samples for each channel (assuming mono audio).

**Channel:**

A "channel" in audio refers to an independent stream of audio data. In stereo audio, you typically have two channels: one for the left speaker and one for the right speaker. Each channel represents the variation in air pressure over time for its corresponding speaker.

Mono Audio: If you have a single channel, it's mono audio. The waveform would be a one-dimensional array, and each value in the array represents the amplitude of the audio signal at a specific sample point.

Stereo Audio: If you have two channels, it's stereo audio. The waveform would be a two-dimensional array, where each row represents a channel, and each column represents a sample point in time.



**The second dimension length and the sample rate being different can occur when dealing with variable-length audio signals. Why?**

More explanation is below:

**Sample Rate:**
The sample rate (measured in Hz or samples per second) indicates how many samples are taken in one second to represent the continuous audio signal. Common sample rates include 44.1 kHz (44,100 samples per second) and 48 kHz.

**Second Dimension Length:**

The second dimension length of the waveform tensor represents the number of samples in each channel. If you have an audio signal with a fixed duration, the number of samples is determined by multiplying the duration (in seconds) by the sample rate.

For example, if you have a 10-second audio clip recorded at a sample rate of 44.1 kHz, the expected number of samples would be:

Number of samples = Duration × Sample Rate

Number of samples = Duration×Sample Rate

Number of samples = 10 seconds × 44,100 samples/second = 441,000

Number of samples= 10seconds × 44,100 samples/second = 441,000

However, if the audio clip has a variable length or if there are gaps in the recording, the actual number of samples may be less than what you would expect from the sample rate and duration calculation.

In this specific case, the waveform tensor has a shape of [2, 499968], which means that for each channel, there are 499,968 samples. This could be due to various reasons, such as variable-length recordings or gaps in the audio data.

###### Audio Possible Channels

The number of channels an audio signal can have is not strictly limited, and it depends on the audio format and the capabilities of the recording, playback, or storage system. However, certain common configurations are widely used. Here are some key ones:

1. **Mono (Monaural):** Mono audio consists of a single channel. This is the simplest form, where the same audio signal is played through all speakers.

2. **Stereo (Binaural):** Stereo audio consists of two channels, typically representing left and right speakers. This is the most common configuration for music, movies, and other forms of audio content.

3. **Surround Sound:** Surround sound configurations can have more than two channels. Common surround sound formats include:
   - **5.1 Surround:** Six channels (front left, front center, front right, rear left, rear right, subwoofer).
   - **7.1 Surround:** Eight channels (front left, front center, front right, rear left, rear right, rear surround left, rear surround right, subwoofer).
   - **Dolby Atmos:** Supports object-based audio with a variable number of channels, allowing for immersive three-dimensional sound.

4. **Ambisonics:** Ambisonics is an audio processing and reproduction technique that can involve a large number of channels to capture a three-dimensional sound field. Common configurations include first-order (4 channels), second-order (9 channels), and third-order (16 channels) ambisonics.

5. **Custom Configurations:** In specialized audio production, custom configurations with a high number of channels may be used for specific applications or experiments.

It's important to note that while the theoretical limit is not strictly defined, practical considerations such as storage space, processing power, and the capabilities of playback systems influence the choice of the number of channels in a particular audio setup. Additionally, the choice of the number of channels depends on the specific requirements of the audio content and the desired listening experience.
